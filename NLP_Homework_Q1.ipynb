```python
#Problem: Considerng a+ in the lexicon as a
#In the end run a for loop in all documents - doing it fo ron rn
def classifier(Doc_URL):
  doc = Doc_URL
  file1 = open(doc, "r")
  FileContent = file1.read()

  import re
  first_dialogue=FileContent
  rgx = re.compile("(\w[\w']*\w|\w)")
  s = first_dialogue.lower()
  l=rgx.findall(s)
#   print("words in our list: ",len(l))
#   print(l)

  #Negative
  Neg_lexicon = open("/content/negative-words.txt","r",encoding="ISO-8859-1").read().lower()
  Neg_points = 0

  words = rgx.findall(Neg_lexicon)
#   print("words in doc:",len(words))
  common_words = [word for word in words if word in l]
#   print(common_words)
  Neg_points = len(common_words)
#   print("Negative Points:",Neg_points)


  #Positive
  Pos_lexicon = open("/content/positive-words.txt","r",encoding="ISO-8859-1").read().lower()
  Pos_points = 0

  words = rgx.findall(Pos_lexicon)
#   print("words in doc:",len(words))
  common_words = [word for word in words if word in l]
#   print(common_words)
  Pos_points = len(common_words)
#   print("Positive Points:",Pos_points)

  Total_points=Pos_points-Neg_points
  if Total_points>=0: 
    return 1
  else: 
    return -1
```

```python
directory1 = '/content/test_folder'
df2=pd.DataFrame()

for filename in os.listdir(directory1):
    # print(filename)
    Pos_neg_ind = classifier('/content/test_folder/'+filename)
    # print(Pos_neg_ind)
    df2=df2.append({'Doc_URL': filename, 'Pred_Labels':Pos_neg_ind}, ignore_index = True)
    result=df.merge(df2, how='inner', on='Doc_URL')

print(result)
```

```python
import numpy as np
accuracy = np.sum(np.equal(result['Label'], result['Pred_Labels'])) / len(result['Label'])
print(accuracy)
```

```python
def f1(actual, predicted):


    # F1 = 2 * (precision * recall) / (precision + recall)
    tp = np.sum((actual==1) & (predicted==actual))
    fp = np.sum((actual==-1) & (predicted==1))
    fn = np.sum((predicted==-1) & (actual==1))
    
    precision = tp/(tp+fp)
    recall = tp/(tp+fn)
    f1 = 2 * (precision * recall) / (precision + recall)
    return f1

f1(result['Label'],result['Pred_Labels'])
```
